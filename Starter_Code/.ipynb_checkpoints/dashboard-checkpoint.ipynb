# San Francisco Rental Prices Dashboard

In this notebook, you will compile the visualizations from the previous analysis into functions that can be used for a Panel dashboard.

```python
# imports
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Read the Mapbox API key
load_dotenv()
mapbox_token = os.getenv("MAPBOX_API_KEY")
```

# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/sfo_neighborhoods_census_data.csv")
sfo_data = pd.read_csv(file_path, index_col="year")

file_path = Path("Data/neighborhoods_coordinates.csv")
df_neighborhood_locations = pd.read_csv(file_path)
```

- - -

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python
# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""
    
    # YOUR CODE HERE!
    mean_units = sfo_data['housing_units'].groupby(['year']).mean()
    mean_units
    mean_units_df = pd.DataFrame(mean_units)
    mean_units_df
    
    chart_min = mean_units_df['housing_units'].min() - mean_units_df['housing_units'].std()
    chart_max = mean_units_df['housing_units'].max() + mean_units_df['housing_units'].std()

    return mean_units_df.hvplot.bar(title='Housing Units in San Francisco from 2010 to 2016',
                            ylim=(chart_min,chart_max))

def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    
    # YOUR CODE HERE!
    avg_rent_salepx_df = sfo_data[['sale_price_sqr_foot','gross_rent']].groupby(['year']).mean()
    
    return avg_rent_salepx_df['gross_rent'].hvplot.bar(title="Average Gross Rent in San Francisco")

def average_sales_price():
    """Average Sales Price Per Year."""
    
    # YOUR CODE HERE!
    avg_rent_salepx_df = sfo_data[['sale_price_sqr_foot','gross_rent']].groupby(['year']).mean()
    
    return avg_rent_salepx_df['sale_price_sqr_foot'].hvplot.line(title="Average Sales Price per Square Foot in San Francisco")

def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""
    
    # YOUR CODE HERE!
    new_sfo_data = sfo_data.groupby(['year','neighborhood']).mean()
    new_sfo_data.reset_index(inplace=True)
    
    return new_sfo_data.hvplot.line(
            x="year",
            y="sale_price_sqr_foot",
            xlabel="year",
            ylabel="avg price per sq ft",
            groupby='neighborhood')

def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""
    
    # YOUR CODE HERE!
    neighborhoods_df = sfo_data.groupby(['neighborhood']).mean()
    top_10_neighborhoods = neighborhoods_df.sort_values('sale_price_sqr_foot',ascending=False)
    top_10_neighborhoods = top_10_neighborhoods.iloc[0:10].reset_index()
    
    return top_10_neighborhoods.hvplot.bar(
            x="neighborhood",
            y="sale_price_sqr_foot",
            xlabel="neighborhood",
            ylabel="avg price per sq ft",
            rot=45)

def parallel_coordinates():
    """Parallel Coordinates Plot."""
    
    # YOUR CODE HERE!
    neighborhoods_df = sfo_data.groupby(['neighborhood']).mean()
    top_10_neighborhoods = neighborhoods_df.sort_values('sale_price_sqr_foot',ascending=False)
    top_10_neighborhoods = top_10_neighborhoods.iloc[0:10].reset_index()
    
    return px.parallel_coordinates(top_10_neighborhoods, color='sale_price_sqr_foot')

def parallel_categories():
    """Parallel Categories Plot."""
    
    # YOUR CODE HERE!
    neighborhoods_df = sfo_data.groupby(['neighborhood']).mean()
    top_10_neighborhoods = neighborhoods_df.sort_values('sale_price_sqr_foot',ascending=False)
    top_10_neighborhoods = top_10_neighborhoods.iloc[0:10].reset_index()
    
    return px.parallel_categories(top_10_neighborhoods,
        dimensions=["neighborhood","sale_price_sqr_foot","housing_units", "gross_rent"],
        color="sale_price_sqr_foot",
        color_continuous_scale=px.colors.sequential.Inferno
        )

def neighborhood_map():
    """Neighborhood Map"""
    
    # YOUR CODE HERE!
    neighborhoods_df = sfo_data.groupby(['neighborhood']).mean()
    neighborhoods_df.reset_index(inplace=True)
    neighborhoods_df.set_index('neighborhood',inplace=True)
    
    join_df = neighborhoods_df.join(df_neighborhood_locations.set_index("Neighborhood")).dropna()
    join_df.reset_index(inplace=True)
    
    px.set_mapbox_access_token(mapbox_token)

    return px.scatter_mapbox(
            join_df,
            lat="Lat",
            lon="Lon",
            size="sale_price_sqr_foot",
            color="gross_rent",
            zoom=12
        )
    
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# YOUR CODE HERE!
yearly_row = pn.Row(housing_units_per_year, average_gross_rent, average_sales_price)
neighborhood_column = pn.Column(average_price_by_neighborhood, top_most_expensive_neighborhoods)
parallel_row = pn.Row(parallel_coordinates, parallel_categories)
welcome_column = pn.Column("Welcome to the San Francisco real estate analysis dashboard. Navigate through the tabs to see information about housing units, rent and price over the last decade broken down by neighborhood.", neighborhood_map)
tabs = pn.Tabs(
    ("Welcome", welcome_column),
    ("Yearly Market Analysis", yearly_row),
    ("Neighborhood Analysis", neighborhood_column),
    ("Parallel Plots Analysis", parallel_row))

panel = pn.Column("#San Francisco Real Estate Analysis Dashboard",tabs)
```

## Serve the Panel Dashboard

```python
panel.servable()
```
